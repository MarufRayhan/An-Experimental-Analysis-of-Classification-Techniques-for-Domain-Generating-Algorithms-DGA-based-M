```python
def apply_cleaning_function_to_list(X):
    cleaned_X = []
    for element in X:
        cleaned_X.append(get_ngrams(element,1))
    return cleaned_X
```

```python
X_train, X_test, y_train, y_test = train_test_split(my_file['clean_url'], 
                                                    my_file['Class'], 
                                                    test_size=0.20, 
                                                    random_state=8)
```

```python
y_train.shape, X_train.shape
```

```python
tfidf = TfidfVectorizer(analyzer='char', token_pattern=r'\w{1,}', ngram_range=(1,4),max_features=5000)
# tfidf.fit(my_file['clean_url'])
tfidf.fit(X_train)
features_train =  tfidf.transform(X_train)
labels_train = y_train

features_test = tfidf.transform(X_test)

labels_test = y_test
# xvalid_tfidf =  tfidf_vect.transform(valid_x)
features_test.data
```
